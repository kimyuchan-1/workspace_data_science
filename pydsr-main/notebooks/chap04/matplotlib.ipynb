```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu, sigma = 0, 1
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 20)
y = norm.pdf(x, mu, sigma)
plt.scatter(x, y, marker='x', s=50, cmap='viridis', alpha=0.3)
plt.colorbar()
plt.title('Normal Distribution')
```
