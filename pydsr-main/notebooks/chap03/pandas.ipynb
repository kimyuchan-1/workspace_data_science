```python
df = pd.read_csv("../data/births.csv")
df.head()
```

```python
df.info()
```

```python
df[df['day'].isna()]
```

```python
df['day'] = df['day'].fillna(1)
df.info()
```

```python
df['day'] = df['day'].astype(int)
```

```python
df['decade'] = 10 * (df['year'] // 10)
```

```python
df.pivot_table('births',index='decade', columns='gender', aggfunc='sum')
```

```python
import matplotlib.pyplot as plt
sns.set()
df.pivot_table('births', index='year', columns='gender', aggfunc='sum').plot()
plt.ylabel('total births per year')
```

```python
import numpy as np
quartiles = np.percentile(df['births'], [25, 50, 75])
mu = quartiles[1]
sig = 0.74 * (quartiles[2] - quartiles[0])
```

```python
df = df.query('(births > @mu - 5 * @sig) & (births < @mu + 5*@sig)')
```

```python
df['day'] = df['day'].astype(int)
```

```python
month = df['month'].astype(str).str.zfill(2)
month
```

```python
df.index = pd.to_datetime(10000 * df.year + 100 * df.month + df.day, format='%Y%m%d')
df['dayofweek'] = df.index.dayofweek
```

```python
import matplotlib as mpl

df.pivot_table('births', index='dayofweek', columns='decade', aggfunc='mean').plot()
plt.gca().set_xticklabels(['Mon','Tues','Wed','Thurs','Fri','Sat','Sun'])
plt.ylabel('mean births by day')
```

```python
births_by_date = df.pivot_table('births', [df.index.month, df.index.day])
births_by_date.head()
```

```python
births_by_date.index = [pd.Timestamp(2012, month, day) for (month, day) in births_by_date.index]
```

```python
births_by_date.head()
```

```python
fig, ax = plt.subplots(figsize=(12,4))
births_by_date.plot(ax=ax)
```

(1) 문자열 alias (가장 흔하게 쓰는 형태)

예시:

'D' → 일 단위 (Day)

'H' → 시간 단위 (Hour)

'T' 또는 'min' → 분 단위 (Minute)

'S' → 초 단위 (Second)

'MS' → 월의 시작 (Month Start)

'M' → 월의 끝 (Month End)

'Q' → 분기 끝 (Quarter End)

'QS' → 분기 시작 (Quarter Start)

'A' 또는 'Y' → 연말 (연도 끝)

'AS' 또는 'YS' → 연초 (연도 시작)

'B' → 영업일(Business day)

'W' → 주간 (기준 요일은 기본 일요일, 'W-MON'처럼 지정 가능)

```python
df = pd.read_csv('../data/FremontBridge.csv', index_col='Date', parse_dates=True)
df.head()
```

```python
df.columns = ['Total','West','East']
```

```python
df.dropna().describe()
```

```python
df.plot()
plt.ylabel('Hourly Bicycle Count')
```

```python
weekly = df.resample('W').sum()
weekly.plot(style=[':','--','-'])
plt.ylabel('Weekly Bicycle Count')
```

```python
daily = df.resample('D').sum()
daily.rolling(30, center = True).sum().plot(style=[':','--','-'])
plt.ylabel('mean hourly Bicycle Count')
```

```python
daily.rolling(50, center=True, win_type='gaussian').sum(std=10).plot(style=[':','--','-'])
```

```python
by_time = df.groupby(df.index.time).mean()
hourly_ticks = 4 * 60 *60 * np.arange(6)
by_time.plot(xticks=hourly_ticks,style=[':','--','-'])
```

```python
by_weekly = df.groupby(df.index.dayofweek).mean()
by_weekly.index = ['Mon','Tues','Wed','Thurs','Fri','Sat','Sun']
by_weekly.plot(style=[':','--','-'])
```

```python
weekend = np.where(df.index.weekday<5,'Weekday','Weekend')
by_time = df.groupby([weekend, df.index.time]).mean()
```

```python
fig, ax = plt.subplots(1,2,figsize=(14,5))
by_time.loc['Weekday'].plot(ax=ax[0], title = 'Weekdays', xticks=hourly_ticks, style=[':','--','-'])
by_time.loc['Weekend'].plot(ax=ax[1], title = 'Weekdends', xticks=hourly_ticks, style=[':','--','-'])
```
